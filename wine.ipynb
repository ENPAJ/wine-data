Question 11 : Créer une colonne "Bon vin" avec la valeur `0` si le vin a une qualité strictement inférieure à 6 et `1` si la valeur est supérieure ou égale à 6. Afficher le résultat dans un graphique en donut avec une légende et un titre.

```python
import matplotlib.pyplot as plt

# Créer la colonne "Bon vin"
df_cleaned['Bon vin'] = df_cleaned['quality'].apply(lambda x: 1 if x >= 6 else 0)

# Compter le nombre de vins "Bon vin" et "Mauvais vin"
counts = df_cleaned['Bon vin'].value_counts()

# Créer un graphique en donut
fig, ax = plt.subplots()
ax.pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightcoral'],
       wedgeprops=dict(width=0.3))

# Ajouter une légende et un titre
ax.legend(['Mauvais vin', 'Bon vin'], loc='upper right')
plt.title('Répartition des vins en fonction de la qualité')

# Afficher le graphique
plt.show()
```

Question 20: Apppliquer un MinMaxScaler sur chaque colonne numérique du dataset puis afficher les nouveaux min, max et moyennes de ces colonnes.

Question 7: Réaliser une matrice de correlation des colonnes finissant par `_se` sous la forme d'une heatmap ou pairplot.

```python
# Sélectionner les colonnes se terminant par "_se"
colonnes_se = [col for col in df.columns if col.endswith('_se')]

# Créer une sous-dataframe avec ces colonnes
df_se = df[colonnes_se]

# Calculer la matrice de corrélation
matrice_correlation_se = df_se.corr()

# Créer une heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(matrice_correlation_se, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Matrice de Corrélation des Colonnes se terminant par "_se"')
plt.show()
```
